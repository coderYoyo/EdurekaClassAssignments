```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("/content/autodata.csv")
df.head()
```

```python
df.replace("?",np.nan,inplace=True)
```

```python
df.head()
```

```python
df.isnull().sum().sum()
```

```python
df.info()
```

```python
df.head()
```

```python
df["peak-rpm"].astype("float").mean()
```

```python
np.array(df['peak-rpm'].dropna().values,dtype="float").mean()
```

```python
df["peak-rpm"].fillna(df["peak-rpm"].astype("float").mean(),inplace=True)
df.head()
```

```python
def find_mean(data):
  x = data.astype("float").mean()
  return x
```

```python
df["curb-weight"].fillna(find_mean(df["curb-weight"]),inplace=True)
df.head()
```

```python
df["wheel-base"].fillna(find_mean(df["wheel-base"]),inplace=True)
df.head()
```

```python
df["body-style"].fillna(df["body-style"].mode()[0],inplace=True) # mode
```

```python
df.head()
```

```python
df["peak-rpm"] = df["peak-rpm"].astype("float")
df["curb-weight"] = df["curb-weight"].astype("float")
df["wheel-base"] = df["wheel-base"].astype("float")
df["horsepower"] = df["horsepower"].astype("float")
df["price"] = df["price"].astype("float")
```

```python
df.boxplot()
```

```python
# Pandas allows you to get quantiles using quantile function
Q1 = df["curb-weight"].quantile(.25)
Q3 = df["curb-weight"].quantile(.75)
IQR = Q3 - Q1
print('The IQR value for curb-weight is',IQR)
```

```python
#remove outliers
# Pandas allows you to get quantiles using quantile function
Q1 = df["price"].quantile(.25)
Q3 = df["price"].quantile(.75)
IQR = Q3 - Q1
print('The IQR value for price is',IQR)
# Let's see which entries are outliers if price column is considered
outliers=df[(df.price < (Q1 - 1.5 * IQR)) |(df.price > (Q3 + 1.5 * IQR))]
outliers
```

```python
#remove outliers
# Pandas allows you to get quantiles using quantile function
Q1 = df["horsepower"].quantile(.25)
Q3 = df["horsepower"].quantile(.75)
IQR = Q3 - Q1
print('The IQR value for horsepower is',IQR)

# Let's see which entries are outliers if horsepower column is considered
outliers=df[(df.horsepower < (Q1 - 1.5 * IQR)) |(df.horsepower > (Q3 + 1.5 * IQR))]
outliers
```

```python
# Pandas allows you to get quantiles using quantile function
Q1 = df["wheel-base"].quantile(.25)
Q3 = df["wheel-base"].quantile(.75)
IQR = Q3 - Q1
print('The IQR value for wheel-base is',IQR)
outliers=df[(df["wheel-base"] < (Q1 - 1.5 * IQR)) |(df["wheel-base"] > (Q3 + 1.5 * IQR))]
outliers
```
